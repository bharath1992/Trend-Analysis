```python
def cleaning(text):

    import string
    exclude = set(string.punctuation)
    import re
    #split attached words
    text = " ".join(re.findall('[A-Z][^A-Z]*', text))
    # remove patterns matching url format
    url_pattern = r'((http|ftp|https):\/\/)?[\w\-_]+(\.[\w\-_]+)+([\w\-\.,@?^=%&amp;:/~\+#]*[\w\-\@?^=%&amp;/~\+#])?'
    text = re.sub(url_pattern, ' ', text)
    # remove hex codes
    text = re.sub(r'(\\x(.){2})', '',text)
    text = nltk.word_tokenize(text)
    text = replace_contractions(text)
    text = remove_non_ascii(text)
    #text = to_lowercase(text)
    text = remove_punctuation(text)
    text = remove_stopwords(text)
    text = remove_small_words(text)
    text = remove_numbers(text)
    return text
```

```python
tweets["Tweet Text"].head()
```

```python
cleaned_tweets = []
for terms in tweets["Tweet Text"]:
    terms = cleaning(terms)
    cleaned_tweets.append(terms)
tweets["Tweet Text"] = [' '.join(line) for line in cleaned_tweets]
```

```python
#Filtering out the GPEs
geography_list = []
for item in tweets["Tweet Text"]:
    doc = nlp(item)
    geography_list.append([(X.text, X.label_)for X in doc.ents if X.label_ == 'GPE' ])
geo_list = list(filter(None, geography_list))
geo_list
```

```python
# Replacing empty list with empty string
geography_list= [x[0] if len(x)!=0 else '' for x in geography_list] 
tweets["GPE"]=geography_list
```
